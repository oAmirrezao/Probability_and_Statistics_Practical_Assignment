  با توجه به قانون اعداد بزرگ انتظار داریم با بزرگ شدن $n$ میانگین نزدیک به $\mathbb{E}[X]=\frac{۱}{\lambda}=۱۰$ بشود و همانطور که مشاهده می‌کنید خروجی قسمت ب) بسیار دقیق تر از قسمت الف) می‌باشد.

    پ) (۱۰ نمره)

<div dir = "rtl" style="direction:rtl;line-height:250%;">
	<font face="XB Zar" size=5>
اکنون نمودار همگرایی میانگین نمونه‌ها به امید ریاضی توزیع نمایی برحسب تعدادشان را برای $n=10000$ رسم‌ کنید.</br>
<b>راهنمایی:</b> بردار زیر را تشکیل دهید:
$$\left(
\frac{x_۱}{۱}, \frac{x_۱ + x_۲}{۲}, \frac{x_۱+x_۲+x_۳}{۳},
\frac{x_۱+x_۲+x_۳+x_۴}{۴}, \dots \right) .
$$
برای تشکل این بردار می‌توانید از تابع $\textrm{cumsum()}$ استفاده کنید.
</div>



```r
n <- 10000

lambda <- 1/10

samples <- rexp(n, rate = lambda)

sample_means <- cumsum(samples) / seq_along(samples)

plot(seq_along(sample_means), sample_means, type = "l", xlab = "Number of Samples", ylab = "Sample Mean", main = "Convergence of Sample Mean to Exponential Distribution Mean")
abline(h = 1/lambda, col = "red")
```

حال می‌خواهیم این نمودار را برای توزیع‌های دیگر نیز رسم کنیم و سرعت و دقت همگرایی آن‌ها را مشاهده کنیم. در سه بخش بعدی، این نمودار را برای توزیع های نرمال، پواسون و کوشی رسم خواهید کرد.
</div>

```r
num_samples <- 1000

mean <- 12
variance <- 5

samples <- rnorm(num_samples, mean = mean, sd = sqrt(variance))

mean_values <- cumsum(samples) / (1:num_samples)

plot(1:num_samples, mean_values, type = "l", col = "blue", xlab = "Number of samples", ylab = "Mean",
     main = "Convergence Plot for Normal Distribution")
abline(h = mean, col = "red", lty = 2)
legend("topright", legend = c("Sample Mean", "True Mean"), col = c("blue", "red"), lty = c(1, 2))
```

کدتان را چند بار اجرا کنید. چه تفاوتی بین این بخش و دو بخش قبل وجود دارد؟ توضیح دهید چرا این تفاوت وجود دارد؟ (نیاز به نوشتن معادلات نیست و توضیح دلیل نیز کافی می‌باشد)

```r
n <- 1000
m <- 4
location <- 0
scale = 1
n_samples <- 1000
m_values <- c(4, 10)

gamma <- 1
x0 <- 0

calculate_mean_pairs <- function(samples, m) {
  sample_means <- matrix(samples, ncol = m)
  rowMeans(sample_means)
}

generate_and_calculate_means <- function(n, m, gamma, x0) {
  samples <- matrix(rcauchy(n * m, location = x0, scale = gamma), ncol = m)
  sample_means <- apply(samples, 1, mean)
  sample_means
}

for (m in m_values) {
  sample_means <- generate_and_calculate_means(n_samples, m, gamma, x0)

  hist(sample_means, breaks = 30, main = paste("Histogram of Sample Means for m = ", m), xlab = "Sample Mean")

}
```

صد ضربه اول محمدپیام را ۱۰۰۰۰ بار شبیه سازی کنید و
            نمودار احتمال گل شدن i توپ از این صد ضربه را بر حسب مقادیر مختلف i، رسم کنید.
            (۸۰ نمره)

```r
library(ggplot2)
num_shots <- 100
num_simulations <- 10000

sim = function(){
  goals <- 1
  for(shot in 3:num_shots) {
    p_goal <- goals / (shot-1)
    p = runif(1, min=0, max=1)
    if(p < p_goal){
      goals = goals + 1
    }
  }
  return (goals)
}

res = array(dim = num_shots, 0)
for (i in 1:num_simulations) {
  sim_result = sim()
  res[sim_result] = res[sim_result] + 1
}



plot(1:num_shots, res / num_simulations, type = "l", xlab = "number of goals", ylab = "probability")

ggplot(frame, aes(x = 1:num_shots, y = res / num_simulations)) +
  geom_bar(stat = "identity") +
  labs(x = "number of goals", y = "probability")
```

```r
test_simulate_step <- function() {
  squares <- c(0, 1, 0, 0, 0)
  new_squares <- simulate_step(squares)
  if (any(new_squares == 0)) {
    print("Test 1 passed: One neighboring square turned white.")
  } else {
    print("Test 1 failed: No neighboring square turned white.")
  }

  squares <- c(0, 1, 0, 1, 0)
  new_squares <- simulate_step(squares)
  if (any(new_squares == 0)) {
    print("Test 2 passed: One neighboring square turned white.")
  } else {
    print("Test 2 failed: No neighboring square turned white.")
  }
}

test_check_adjacent_whites <- function() {
  squares_1 <- c(1, 1, 1, 1, 1)
  if (check_adjacent_whites(squares_1)) {
    print("Test case 1 failed: No white squares")
  } else {
    print("Test case 1 passed")
  }

  squares_2 <- c(1, 0, 1, 1, 1)
  if (check_adjacent_whites(squares_2)) {
    print("Test case 2 failed: Single white square")
  } else {
    print("Test case 2 passed")
  }

  squares_3 <- c(1, 0, 0, 1, 1)
  if (check_adjacent_whites(squares_3)) {
    print("Test case 3 passed")
  } else {
    print("Test case 3 failed: Adjacent white squares")
  }

  squares_4 <- c(1, 0, 1, 0, 1)
  if (check_adjacent_whites(squares_4)) {
    print("Test case 4 failed: Non-adjacent white squares")
  } else {
    print("Test case 4 passed")
  }
}



test_simulate_process <- function() {
  N <- 3
  result <- simulate_process(N)
  if (result == 1/3) {
    print("Test 1 passed: w(N)/N is correct.")
  } else {
    print("Test 1 failed: Incorrect w(N)/N value.")
  }

  N <- 4
  result <- simulate_process(N)
  if (result == 2/4) {
    print("Test 2 passed: w(N)/N is correct.")
  } else {
    print("Test 2 failed: Incorrect w(N)/N value.")
  }
}

test_simulate_N_range <- function() {
  start_N <- 10
  end_N <- 30
  step_size <- 10
  num_simulations <- 2
  results <- simulate_N_range(start_N, end_N, step_size, num_simulations)
  if (nrow(results) == 6 && ncol(results) == 2) {
    print("Test passed: Correct number of rows and columns in results dataframe.")
  } else {
    print("Test failed: Unexpected number of rows or columns in results dataframe.")
  }
}
```

```r
simulate_step <- function(squares) {
  white_indices <- which(squares == 1)

  sample_square <- sample(white_indices, 1)

  selected_square <- 0
  left_or_right_neighbour <- runif(1, 0, 1)
  if(left_or_right_neighbour > 0.5) {
    selected_square <- (sample_square + 1)
  } else {
    selected_square <- (sample_square - 1)
  }


  if(squares[selected_square] == 1) {
    squares[selected_square] = 0
  }

  return(squares)
}
```

```r
test_simulate_step()
```

```r
check_adjacent_whites <- function(squares) {
  white_indices <- which(squares == 1)
  if(length(white_indices < 2)) {
    return (FALSE)
  }
  for(i in white_indices) {
    if(squares[i-1] + squares[i+1] > 1) { return (TRUE)}
  }
  return (FALSE)
}
```

```r
test <- c(1, 0, 0, 1, 1)
res <- check_adjacent_whites(test)
print(res)
```

```r
test_check_adjacent_whites()
```

```r
simulate_process <- function(N) {
  squares <- c(0, rep(1, N), 0)
  while(check_adjacent_whites(squares)) {
    squares <- simulate_step(squares)
  }
  no_of_remained_white_squares <- which(squares == 1)
  w <- length(no_of_remained_white_squares)
  res <- w / N
  return (res)
}
```

```r
test_simulate_process()
```

```r
simulate_N_range <- function(start_N, end_N, step_size, num_simulations) {
  results <- data.frame(nValue = numeric(), rate = numeric())
  total_number <- seq(start_N, end_N, step_size)
  for(i in total_number) {
    for (j in 1:num_simulations) {
      results <- rbind(results, data.frame(nValue = i, rate = simulate_process(i)))
    }
  }
  return (results)
}
```

```r
library(ggplot2)
plot_results <- function(results) {

  plot<-ggplot(results,aes(nValue,rate))+geom_point(colour="red")+ggtitle("main plot")+xlab("N")+ylab("w(N)/N")+geom_smooth(method = "loess",formula = y ~ x)
  return (plot)
}
```

```r
main <- function(start_N, end_N, step_size, num_simulations) {
  true_value <- 1 / exp(1)

  results <- simulate_N_range(start_N, end_N, step_size, num_simulations)

  mean_calculated_value <- mean(results[["rate"]])

  cat("Calculated Value: ", mean_calculated_value, "\n")
  cat("True Value (1/e): ", true_value, "\n")
  cat("Absolute Error: ", abs(mean_calculated_value - true_value), "\n")

  plot_results(results)
}
```
